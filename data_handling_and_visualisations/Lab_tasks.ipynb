```python
# Create a new DataFrame for each of the species separately - should have 3 in total, make sure each index starts at 0
setosa_df = iris[iris['species'] == 'setosa'].reset_index(drop=True)
versicolor_df = iris[iris['species'] == 'versicolor'].reset_index(drop=True)
virginica_df = iris[iris['species'] == 'virginica'].reset_index(drop=True)

print(setosa_df.head())
print(versicolor_df.head())
print(virginica_df.head())
```

```python
# Create two final DataFrames: One for the Sepal information, and one for the Petal information.
# This should be done on the whole dataset so that in each DataFrame we have information about all three species

sepal_df = iris[['sepal_length','sepal_width','species']].copy()
print(sepal_df)

petal_df = iris[['petal_length','petal_width','species']].copy()
print(petal_df)
```
